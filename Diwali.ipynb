```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("/kaggle/input/diwali-sales-dataset/Diwali Sales Data.csv", encoding='unicode_escape')
df.info()
df.pop('Status')
df.pop('unnamed1')

df.isna().sum()
df.dropna(inplace=True)

#customers w.r.t gender
plt.figure(1)
plt.pie(df['Gender'].value_counts(),labels=['Female', 'male'])
plt.xlabel('Gender distribution')


Male=df[df['Gender']=='M']
plt.figure(2)
plt.pie(Male['Age Group'].value_counts(),labels=['26-35','36-45','18-25','46-50','51-55','55+','0-17'])
plt.xlabel('Male age group distribution')
Male_shop_stat=Male['Amount'].describe()

#female proportion w.r.t age groups

Female=df[df['Gender']== 'F']
plt.figure(3)
plt.pie(Female['Age Group'].value_counts(),labels=['26-35','36-45','18-25','46-50','51-55','55+','0-17'])
plt.xlabel('Female age group distribution')
Female_shop_stat=Female['Amount'].describe()


#Top 100 Sales and States
plt.figure(4)
city=(df.nlargest(100, ['Amount'])['State'].value_counts())
city.plot(kind='bar')
plt.xlabel('States')
plt.ylabel('Occurances')
plt.title('States where top 100 sales occured')


#Age wise distribution
plt.figure(5)
plt.hist(df['Amount'], bins=30)
plt.title('age-wise distribution')
plt.figure(6)
#Age Group wise distribution
plt.hist(df['Age Group'])
plt.title('age group wise distribution')


# state wise distribution
plt.figure(7)
plt.hist(df['State'])
plt.xticks(rotation='vertical')
plt.xlabel('State')
plt.ylabel('Occurances')

# which state had higest sales
plt.figure(8)
df2=df.groupby('State')['Amount'].sum().plot(kind='bar')

#make a groupby based on states and products bought
zz=df.groupby('State')['Product_Category']

c=10
cityname=df['State'].unique()
for name in cityname:
    plt.figure(c)
    zz.get_group(name).value_counts().plot(kind='bar')
    plt.title(f'Items sold in {name} for diwali sales')
    c=c+1



```
